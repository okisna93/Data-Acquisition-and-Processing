```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import statsmodels.api as sm
from statsmodels.tsa.arima.model import ARIMA


DATA_PATH_Q2 = Path.cwd() / 'EE627A_HW2_Q2.csv'
DATA_PATH_Q3 = Path.cwd() / 'EE627A_HW2_Q3.csv'
```

#  Find the best predictive model for Q2

```python
Q2_df = pd.read_csv(DATA_PATH_Q2, names=["Q2"])
Q2_df.head()
```

```python
# Check for null values
Q2_df.isnull().values.any()
```

```python
# Plot data
Q2_df['Q2'].plot()
```

### ACF for Q2

```python
plot_acf(Q2_df["Q2"]);
```

### PACF for Q2

```python
plot_pacf(Q2_df["Q2"]);
```

Looks like MA(2) according to ACF suddenly hitting close to 0 after 2.

```python
ma2_model = ARIMA(Q2_df['Q2'], order=(0,0,2))
ma2_model_fit = ma2_model.fit()
ma2_model_fit.summary()
```

```python
plot_acf(ma2_model_fit.resid);
```

```python
plot_pacf(ma2_model_fit.resid);
```

Residuals are not all close to 0, so we will try MA(3)

```python
ma3_model = ARIMA(Q2_df['Q2'], order=(0,0,3))
ma3_model_fit = ma3_model.fit()
ma3_model_fit.summary()
```

```python
plot_acf(ma3_model_fit.resid);
```

```python
plot_pacf(ma3_model_fit.resid);
```

MA(3) is good enough for this time series called Q2.

#  Find the best predictive model for Q3

```python
Q3_df = pd.read_csv(DATA_PATH_Q3, names=["Q3"])
Q3_df.head()
```

```python
# Check for null values
Q3_df.isnull().values.any()
```

```python
# plot
Q3_df['Q3'].plot()
```

### ACF for Q3

```python
plot_acf(Q3_df["Q3"]);
```

### PACF for Q3

```python
plot_pacf(Q3_df["Q3"]);
```

The data doesn't look stationary and the ACF and PACF aren't reliable until the data is stationary so we will take the first order difference

```python
Q3_df['Q3_1st_order_diff'] = Q3_df['Q3'] - Q3_df['Q3'].shift(1)
Q3_df.head()
```

```python
first_order_diff = Q3_df['Q3_1st_order_diff'][1:]
first_order_diff.plot()
```

### ACF for 1st order difference

```python
plot_acf(first_order_diff);
```

### Plot PACF for 1st order difference

```python
plot_pacf(first_order_diff);
```

```python
ar1ma1_model = ARIMA(first_order_diff, order=(1,1,1))
ar1ma1_model_fit = ar1ma1_model.fit()
ar1ma1_model_fit.summary()
```

```python
ar1ma1_model_fit.resid
```

```python
plot_acf(ar1ma1_model_fit.resid[1:]);
```

```python
plot_pacf(ar1ma1_model_fit.resid[1:]);
```

ACF and PACF residuals do not stay consistently close to 0. So we will try ARIMA(2, 1, 1)

```python
ar2ma1_model = ARIMA(first_order_diff, order=(2,1,1))
ar2ma1_model_fit = ar2ma1_model.fit()
ar2ma1_model_fit.summary()
```

```python
plot_acf(ar2ma1_model_fit.resid);
```

```python
plot_pacf(ar2ma1_model_fit.resid);
```

```python
ar3ma1_model = ARIMA(first_order_diff, order=(3,1,1))
ar3ma1_model_fit = ar3ma1_model.fit()
```

```python
plot_acf(ar3ma1_model_fit.resid);
```

```python
plot_pacf(ar3ma1_model_fit.resid, alpha=0.05);
```

```python
sm.graphics.tsa.pacf(ar3ma1_model_fit.resid, alpha=0.05, nlags=30)
```

```python
ar0ma10_model = ARIMA(first_order_diff, order=(0,1,10))
ar0ma10_model_fit = ar0ma10_model.fit()
ar0ma10_model_fit.summary()
```

```python
plot_acf(ar0ma10_model_fit.resid)
```
