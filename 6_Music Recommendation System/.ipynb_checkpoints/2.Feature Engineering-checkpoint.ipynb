```python
import pandas as pd
from sklearn.decomposition import PCA
import statistics
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.decomposition import NMF
```

```python
data=pd.read_csv('8_Features.csv')
data.tail()
```

```python
data=data[['AlbumId','ArtistId','Track','Genre1','Genre2','Genre3','Genre4','Genre5','Genre6']]
data
```

```python
#Feature Engineering

#Summation ---> AlbumId+ArtistId
data['Feature1']=data['AlbumId'].values+data['ArtistId'].values

#Average ---> AlbumId+ArtistId
data['Feature2']=(data['AlbumId'].values+data['ArtistId'].values)/2

#Weighted Average --->  0,7 and 0,3  AlbumId+ArtistId
data['Feature3']=(0.7*data['AlbumId'].values+0.3*data['ArtistId'].values)/2

#Weighted Average --->  0,75 and 0,25  AlbumId+ArtistId
data['Feature4']=(0.75*data['AlbumId'].values+0.25*data['ArtistId'].values)/2

#Weighted Average --->  0,8 and 0,2  AlbumId+ArtistId
data['Feature5']=(0.8*data['AlbumId'].values+0.2*data['ArtistId'].values)/2

#Weighted Average --->  0,85 and 0,15  AlbumId+ArtistId
data['Feature6']=(0.85*data['AlbumId'].values+0.15*data['ArtistId'].values)/2

#Weighted Average --->  0,90 and 0,10  AlbumId+ArtistId
data['Feature7']=(0.90*data['AlbumId'].values+0.10*data['ArtistId'].values)/2

#Weighted Average --->  0,95 and 0,05  AlbumId+ArtistId
data['Feature8']=(0.95*data['AlbumId'].values+0.5*data['ArtistId'].values)/2

#Summation ---> AlbumId+ArtistId+Track
data['Feature9']=data['AlbumId'].values+data['ArtistId'].values+data['Track'].values

#Average ---> AlbumId+ArtistId+Track/3
data['Feature10']=(data['AlbumId'].values+data['ArtistId'].values+data['Track'].values)/3

#Weighted Average ---> 0,6 and 0,3 and 0,1 AlbumId+ArtistId+Track/3
data['Feature11']=(0.6*data['AlbumId'].values+0.3*data['ArtistId'].values+0.1*data['Track'].values)/3

#Weighted Average ---> 0,7 and 0,2 and 0,1 AlbumId+ArtistId+Track/3
data['Feature12']=(0.6*data['AlbumId'].values+0.2*data['ArtistId'].values+0.1*data['Track'].values)/3

#Weighted Average ---> 0,8 and 0,15 and 0,05 AlbumId+ArtistId+Track/3
data['Feature13']=(0.8*data['AlbumId'].values+0.15*data['ArtistId'].values+0.05*data['Track'].values)/3

#Summation ---> AlbumId+ArtistId+Track+Genre1
data['Feature14']=data['AlbumId'].values+data['ArtistId'].values+data['Track'].values+data['Genre1'].values

#Average ---> AlbumId+ArtistId+Track+Genre1/4
data['Feature15']=(data['AlbumId'].values+data['ArtistId'].values+data['Track'].values+data['Genre1'].values)/4

#Summation ---> AlbumId+ArtistId+Track+Genre1+Genre2
data['Feature16']=data['AlbumId'].values+data['ArtistId'].values+data['Track'].values+data['Genre1'].values+data['Genre2'].values

#Average ---> (AlbumId+ArtistId+Track+Genre1+Genre2)/5

data['Feature17']=(data['AlbumId'].values+data['ArtistId'].values+data['Track'].values+data['Genre1'].values+data['Genre2'].values)

#Summation ---> AlbumId+ArtistId+Track+Genre1+Genre2+Genre3+Genre4+Genre5+Genre6
data['Feature18']=(data['AlbumId'].values+data['ArtistId'].values+data['Track'].values
                   +data['Genre1'].values+data['Genre2'].values
                   +data['Genre3'].values+data['Genre4'].values
                  +data['Genre5'].values+data['Genre6'].values)

#Average --->(AlbumId+ArtistId+Track+Genre1+Genre2+Genre3+Genre4+Genre5+Genre6)/8

data['Feature19']=(data['AlbumId'].values+data['ArtistId'].values+data['Track'].values
                   +data['Genre1'].values+data['Genre2'].values
                   +data['Genre3'].values+data['Genre4'].values
                  +data['Genre5'].values+data['Genre6'].values)/8

#Standart Mean --->AlbumId+ArtistId+Track+Genre1+Genre2+Genre3+Genre4+Genre5+Genre6
data['Feature20']=data[['AlbumId']].mean(axis=1,skipna=True)

#Standart Deviation --->AlbumId+ArtistId+Track+Genre1+Genre2+Genre3+Genre4+Genre5+Genre6
data
```

```python
data.to_csv('21_Features.csv')
```

```python
pca=PCA(n_components=1)
pca_data=pca.fit_transform(data)
```

```python
pca_dataframe=pd.DataFrame(data=pca_data,columns=['Predictor'])
pca_dataframe.abs()
```

```python
Output=pd.read_csv('OUTPUT3.csv')
Output['Predictor']=pca_dataframe.abs()
Output
```

```python
model = NMF(n_components=1, init='random', random_state=0)
W = model.fit_transform(data)
H = model.components_
decomposition_data=pd.DataFrame(data=W,columns=['Predictor'])
decomposition_data
```

```python
Output2=pd.read_csv('OUTPUT3.csv')
Output2['Predictor']=decomposition_data
Output2
```
