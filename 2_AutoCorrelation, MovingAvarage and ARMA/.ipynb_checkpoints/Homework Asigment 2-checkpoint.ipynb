```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.arima_model import ARIMA
```

## Problem-2

```python
data1_Q2=pd.read_csv('EE627A_HW2_Q2.csv',names=['Values'])
data1_Q2
```

```python
sns.lineplot(data=data1_Q2)
plt.xlabel('Number of Data')
plt.ylabel('Values')
plt.show()
```

Data seems stationary.

```python
plot_acf(data1_Q2, lags=25);
plot_pacf(data1_Q2, lags=25);
```

As we can see on the ACF graph that graph make a cliff after MA(2)

```python
Model_Q2= ARIMA(data1_Q2, order=(0, 0, 2))
Model_Q2_fit =Model_Q2.fit()
Model_Q2_fit.summary()
```

```python
Q2_residual = pd.DataFrame(Model_Q2_fit.resid)
plot_acf(Q2_residual, title='Q2 residuals ACF');
plot_pacf(Q2_residual, title='Q2 residuals PACF');
plt.legend()
plt.show()
```

There are points which are not in the range of 0 that is why  we will try MA(3) model

```python
Model_Q2= ARIMA(data1_Q2, order=(0, 0, 3))
Model_Q2_fit =Model_Q2.fit()
Model_Q2_fit.summary()
```

```python
Q2_residual = pd.DataFrame(Model_Q2_fit.resid)
plot_acf(Q2_residual, title='Q2 residuals ACF');
plot_pacf(Q2_residual, title='Q2 residuals PACF');
plt.legend()
plt.show()
```

### As we can see on the graphs that all the points are close to zero we can say that MA(3) model is good enough for this time series

## Problem-3

```python
data1_Q3=pd.read_csv('EE627A_HW2_Q3.csv',names=['Values'])
data1_Q3
```

```python
sns.lineplot(data=data1_Q3)
plt.xlabel('Number of Data')
plt.ylabel('Values')
plt.show()
```

As we can see on the graph that time series is not Stationary

```python
plot_acf(data1_Q3, lags=25);
plot_pacf(data1_Q3, lags=25);
```

### In order to make the time series Stationary we will take first order difference

```python
#Taking Difference in order to Make Stationary
difference1=data1_Q3['Values'].diff(periods=1)
difference1=difference1[1:]
difference1.head()
```

```python
#Plotting Data After First Order Difference
sns.lineplot(data=difference1)
plt.xlabel('Number of Data')
plt.ylabel('Values')
plt.show()
```

### After taking first order difference data seems stationary.

```python
plot_acf(difference1, lags=25);
plot_pacf(difference1, lags=25);
```

```python
Model_Q3_1= ARIMA(difference1, order=(4, 1, 4))
Model_Q3_1_fit =Model_Q3_1.fit()
Model_Q3_1_fit.summary()
```

```python
Q3_residuals = pd.DataFrame(Model_Q3_1_fit.resid)
plot_acf(Q3_residuals, title='Q3-1 Residuals ACF');
plot_pacf(Q3_residuals, title='Q3-1 Residuals PACF');
plt.show()
```

### After, I had tried many different models ((2,1,2),(2,1,4),(5,1,5),(3,1,4)...)  ARIMA(4,1,4) is the best model I have achieved
